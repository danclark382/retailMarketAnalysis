```python
import glob
import pandas as pd
```

### Calculating daily percent change in stock price

```python
files = glob.glob("data/performance/*.csv")
for f in files:
    print(f)
    try:
        df = pd.read_csv(f, index_col='date', parse_dates=True)
    except ValueError:
        print(f)
        continue
    daily_pc = df.close/df.close.shift(1) - 1
    df['daily_pct'] = daily_pc
    df.to_csv(f)
```

### Calculating daily percent change in user ownership and aggregating

```python
files = glob.glob("data/popularity_export/*.csv")
for f in files:
    try:
        df = pd.read_csv(f, index_col='timestamp', parse_dates=True)
    except ValueError:
        print(f)
        continue    
    df2 = df.resample('D').agg({'users_holding': 'last'})
    daily_pc2 = df2.users_holding/df2.users_holding.shift(1) - 1
    df2['daily_pct'] = daily_pc2
    df2.to_csv("data/popularity_export/daily/" + f.split('/')[-1])
    
```

### Merge all dataframes join on date index for aggregation

```python
start = glob.glob("data/popularity_export/daily/*.csv")[0]
tick = start.split('/')[-1].split('.')[0]
df = pd.read_csv(start, index_col='timestamp', parse_dates=True, usecols=['timestamp', 'users_holding']).rename(columns={'users_holding': tick})
for f in glob.glob("data/popularity_export/daily/*.csv")[0:]:
    tick = f.split('/')[-1].split('.')[0]
    df2 = pd.read_csv(f, index_col='timestamp', parse_dates=True, usecols=['timestamp', 'users_holding']).rename(columns={'users_holding': tick})
    df = pd.merge(df, df2, how='outer', left_index=True, right_index=True)
```

```python
df["sum"] = df.sum(axis=1)
df.to_csv("merged_popularity.csv")
```
